### IMPORT LIBRARIES

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, sum
```

### Create Spark Session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("BigMart Data Analysis").getOrCreate()
```

### Load Dataset


```python
df1 = spark.read.format('csv') \
    .option('inferSchema', True) \
    .option('header', True) \
    .load('/Workspace/Users/banerjeenil027@gmail.com/Databricks tutorials/BigMart Sales.csv')
   
```

```python
df1.display()
```

### Explore Dataset

```python
df1.printSchema()
```

```python
df1.describe().display()
```

###CHECK THE MISSING VALUE 

```python

from pyspark.sql.functions import col, sum

df1.select([sum(col(c).isNull().cast('int')).alias(c) for c in df1.columns]).display()
```

### Data Cleaning

```python
df1 = df1.fillna({
    "Item_Weight":0,
    "Outlet_Size":"Unknown"
})

df1.display()
```

#### Data Analysis
#### Total Sales by Outlet Type

```python
from pyspark.sql.functions import sum, col, round

df1.groupBy("Outlet_Type") \
   .agg(round(sum("Item_Outlet_Sales"), 2).alias("Total_Sales")) \
   .display()
```

### Average Sales by Item Type

```python
df1.groupBy('Item_Type').agg(round(avg('Item_Outlet_Sales') , 2).alias('Average_Sales')).display()
```

### Sales by Location Type

```python
df1.groupBy("Outlet_Location_Type").agg(round(sum("Item_Outlet_Sales") , 2).alias("Sales by Location Type")).show()
```

### Top Selling Products

```python
df1.orderBy(col("Item_Outlet_Sales").desc())
df1.limit(10).display()
```

### “Supermarket Type outlets generate the highest sales.”

```python
df1.groupBy("Outlet_Type") \
   .agg(sum("Item_Outlet_Sales").alias("Total_Sales")) \
   .orderBy(col("Total_Sales").desc()) \
   .display()
```

### Tier-3 locations have higher sales compared to Tier-1 and Tier-2 locations.

```python
df1.groupBy("Outlet_Location_Type") \
   .agg(sum("Item_Outlet_Sales").alias("Total_Sales")) \
   .orderBy(col("Total_Sales").desc()) \
   .display()
```

### Some item categories such as Fruits, Snacks, and Household items show strong sales performance

```python
df1.groupBy("Item_Type") \
   .agg(sum("Item_Outlet_Sales").alias("Total_Sales")) \
   .orderBy(col("Total_Sales").desc()) \
   .display()
```

#### Products with higher MRP tend to generate higher sales revenue.

```python
df1.groupBy("Item_MRP") \
   .agg(avg("Item_Outlet_Sales").alias("Avg_Sales")) \
   .orderBy("Item_MRP") \
   .display()
```
